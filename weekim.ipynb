```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
csv1=pd.read_csv('week2.csv')
csv1=csv1.dropna(0)
csv1.head()
```

```
csv1.info()
```

```
csv1['Date'] =  pd.to_datetime(csv1['Date'], infer_datetime_format=True)
csv1.info()
```

```
csv1.set_index('Date',inplace=True)
csv1['Close Price'].plot(figsize=(16,12))
```

```
plt.figure(figsize=(12,6))

plt.stem(csv1['year'],
         csv1['Day_Perc_Change'],
         markerfmt = 'b_',
         linefmt = 'r--',
         basefmt = 'g:'
)
plt.show()

```
