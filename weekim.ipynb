```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
csv1=pd.read_csv('week2.csv')
csv1=csv1.fillna(0)
csv1.head()
```

```
csv1.info()
```

```
csv1['Date'] =  pd.to_datetime(csv1['Date'], infer_datetime_format=True)
csv1.info()
```

```
csv1.set_index('Date',inplace=True)
csv1['Close Price'].plot(figsize=(16,12))
```

```
plt.figure(figsize=(12,6))

plt.stem(csv1['year'],
         csv1['Day_Perc_Change'],
         markerfmt = 'b_',
         linefmt = 'r--',
         basefmt = 'g:'
)
plt.show()

```

```
csv1['vwap'].plot(figsize=(12,6))
```

```
csv1=pd.groupby(by='Trends',group_keys=False)
figureObject, axesObject = plt.subplots()
axesObject.pie(csv1['Day_Perc_Change'],

        labels=csv1['Trends'],

        autopct='%1.2f',

        startangle=90)
axesObject.axis('equal')
plt.show()



```
