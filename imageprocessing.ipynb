```
from skimage import data
import numpy as np

import matplotlib.pyplot as plt
rocket_im=data.rocket()
def show_image (image,title='Image',camp_type='gray'):
  plt.imshow(image)
  plt.title(title)
  plt.axis('Off')
  plt.show()
show_image(rocket_im)
```

```
from skimage import color
grayscale=color.rgb2gray(rocket_im)
show_image(grayscale)
```

```
verticlly_flipping=np.flipud(rocket_im)
show_image(verticlly_flipping)
```

```
horzontally_flipping=np.fliplr(rocket_im)
show_image(horzontally_flipping)
```

```
red_channel=rocket_im[:,:,0]
plt.hist(red_channel.ravel(),bins=256)
plt.show()
```

```
blu_channel= rocket_im[:,:,1]
plt.hist(blu_channel.ravel(),bins=256)
plt.show()
```

```
from skimage.filters import threshold_otsu
from skimage.color import rgb2gray
colorcha=rgb2gray(rocket_im)
thresh= threshold_otsu(rocket_im)
binary_thresh= colorcha > thresh
show_image(binary_thresh)
```
