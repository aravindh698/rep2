Load data using skimage and a function for showing image

```
from skimage import data
import numpy as np

import matplotlib.pyplot as plt
rocket_im=data.rocket()
def show_image (image,title='Image',camp_type='gray'):
  plt.imshow(image)
  plt.title(title)
  plt.axis('Off')
  plt.show()
show_image(rocket_im)
```

Converting the image to grayscale

```
from skimage import color
grayscale=color.rgb2gray(rocket_im)
show_image(grayscale)
```

Flipping -vetically

```
verticlly_flipping=np.flipud(rocket_im)
show_image(verticlly_flipping)
```

Flipping -horizontally

```
horzontally_flipping=np.fliplr(rocket_im)
show_image(horzontally_flipping)
```

```
red_channel=rocket_im[:,:,0]
plt.hist(red_channel.ravel(),bins=256)
plt.show()
```

```
blu_channel= rocket_im[:,:,1]
plt.hist(blu_channel.ravel(),bins=256)
plt.show()
```

```
from skimage.filters import threshold_otsu
from skimage.color import rgb2gray
colorcha=rgb2gray(rocket_im)
thresh= threshold_otsu(rocket_im)
binary_thresh= colorcha > thresh
show_image(binary_thresh)
```

```
from skimage.filters import gaussian
gaussian_image=gaussian(rocket_im,multichannel=True)
show_image(gaussian_image)
```

EQUALIZATION 

```
from skimage import exposure
image_eq=exposure.equalize_hist(rocket_im)
show_image(image_eq)
```

```
imge_eq=exposure.equalize_adapthist(rocket_im,clip_limit=0.03)
show_image(imge_eq)
```

rotate left

```
from skimage.transform import rotate
image_rotate=rotate(rocket_im,90)
show_image(image_rotate)
```

Rotate right

```
image_rotate=rotate(rocket_im,-90)
show_image(image_rotate)
```

```
# Import the module and function to enlarge images
from skimage.transform import rescale

# Import the data module
from skimage import data

# Load the image from data
rocket_image = data.rocket()

# Enlarge the image so it is 3 times bigger
enlarged_rocket_image = rescale(rocket_image,1/3,anti_aliasing=True, multichannel=True)

# Show original and resulting image
show_image(rocket_image)
show_image(enlarged_rocket_image, "3 times enlarged image")
```

```
# Import the module and function
from skimage.transform import resize

# Set proportional height so its half its size
height = int(rocket_im.shape[0] / 2)
width = int(rocket_im.shape[1] / 2)

# Resize using the calculated proportional height and width
image_resized = resize(rocket_im, (height,width),
                       anti_aliasing=True)

# Show the original and rotated image
show_image(rocket_im, 'Original')
show_image(image_resized, 'Resized image')
```
